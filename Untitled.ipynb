```python
link="https://github.com/LUCEROBOCANEGRA/prac3/raw/main/appendData/"
#file names a list
level0=range(1,5)
filename=['level'+str(level)+'.xlsx' for level in level0]
filename
```

```python
#creamos el url
links=[link+fn for fn in filename]
links
```

```python
#crar listas de df
import pandas as pd
dfs=[pd.read_excel(link) for link in links]
```

```python
#emos cuantos hay
len(dfs)
```

```python
#leemos los dfs, vemos q el primero inicia desde la fila 4 para leer los encabezados asi que esa la hacemos leer desde ahí
dfs = []
dfs.append(pd.read_excel(links[0], header=4))

# Leer los demás archivos con la configuración por defecto
for i in range(1, len(links)): 
    dfs.append(pd.read_excel(links[i]))
```

```python
#para la primera, segunda y tercera tabla, se eliminan las dos primeras filas
for i in range(0, 3): 
    dfs[i] = dfs[i].drop(dfs[i].index[:2])  # Elimina las primeras 2 filas
    dfs[i] = dfs[i].reset_index(drop=True)  # Reinicia el índice
```

```python
dfs[1]
```

```python
dfs[2]
```

```python
# Obtener un df que contenga de todos los dfs las mismams columnas
encabezados_primer_df = list(dfs[0].columns)

# Encontrar los encabezados comunes a todos los DataFrames
encabezados_comunes = encabezados_primer_df.copy()
for df in dfs[1:]:
    encabezados_comunes = [encabezado for encabezado in encabezados_comunes if encabezado in df.columns]

# Seleccionar solo las columnas con encabezados comunes en cada DataFrame
for i in range(len(dfs)):
    dfs[i] = dfs[i][encabezados_comunes]

# Concatenar los DataFrames
df_final = pd.concat(dfs, ignore_index=True)
```

```python
import pandas as pd
enlace ="https://github.com/LUCEROBOCANEGRA/prac3/raw/refs/heads/main/countryCODES.xlsx"
country= pd.read_excel(enlace)
country
```

```python
#como los nombres están en mayúsculas, formateamos el DF inicial
df_final['Country']=df_final.Country.str.upper()
df_final
```

```python
# Realizar el merge. Asumimos que la clave común es 'Country'
merged_df = df_final.merge(country,how='left', left_on='Country', right_on = 'Country')
merged_df
```

```python
#renombrando columnas
newNames = ['GNI', 'rank', 'years_school', 'life_exp', 'GNIminHDI','exp_school', 'HDI', 'country', 'of_name', 'TLD', 'iso2', 'iso3']
merged_df.columns = newNames
merged_df
```

```python
merged_df.isnull().sum()
```

```python

#eliminando los valores:
merged_df = merged_df.dropna()
merged_df
```

```python
#ordenando las columnas
merged_df.set_index(['country', 'of_name', 'rank', 'TLD', 'iso2', 'iso3'], inplace=True)
merged_df.head()
```

```python
#orden alfabético
merged_df.sort_index(axis=1,inplace=True)
```

```python
#regresando los índices seleccionados
merged_df.reset_index(inplace=True)
merged_df
```
